### Scatter

Para el siguiente gráfico "scatter" utizaremos un cmap, los cmap disponibles están en la siguiente página:  
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
#gráfico de burbujas o grafico de bolas
def f(i,j):
    return (j**2)*(i**2)

x=[]
y=[]
z=[]
N=10
step=2
for i in np.arange(-N, N+step, step):
    for j in np.arange(-N, N+step, step):
        x.append(i)
        y.append(j)
        z.append(f(i,j))
        
plt.scatter(x, y, c=z, cmap='viridis',s=z, alpha=0.5)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)
plt.show()
```

Haz la gráfica del analisis de frecuencias realizado del archivo "../04_RW/el_quijote.txt"  
tienes que obtener la siguiente gráfica:  
<img src="f6.png" ALIGN="left">

```python
import matplotlib.pyplot as plt
import numpy as np

archivo=open("../04_RW/el_quijote.txt", encoding='utf8')
text=archivo.readlines()


abc="ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"

dataArray=[]
textArray=[]
continuedText = ""

for line in text:
    
    continuedText += line
    
    continuedText = continuedText.upper()
    continuedText = continuedText.replace('Á','A')
    continuedText = continuedText.replace('É','E')
    continuedText = continuedText.replace('Í','I')
    continuedText = continuedText.replace('Ó','O')
    continuedText = continuedText.replace('Ú','U')
    
for i in continuedText:
    if(i in abc):
            textArray.append(i)
            
count = len(textArray)

for i in abc:
    n = textArray.count(i)
    countVar = n/count
    
    dataArray.append(countVar)


plt.hist(dataArray, edgecolor='black')
plt.show()
```
