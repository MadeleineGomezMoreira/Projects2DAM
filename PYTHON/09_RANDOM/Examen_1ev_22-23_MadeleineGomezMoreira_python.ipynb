Tira tres monedas, realiza el suficiente número de tiradas para responder la siguiente pregunta.
¿Cuál sería  la probabilidad de obtener tres cruces?

```python
#En este ejercicio, cara = 1 y cruz = 0  

numTiradas = 100000
combinaciones = []


def tiradas(numero):
    
    for i in range(numero):
        
        tresMonedas = []
        
        tiroOne = random.randint(0,1)
        tiroTwo = random.randint(0,1)
        tiroThree = random.randint(0,1)
        
        tresMonedas.append(tiroOne)
        tresMonedas.append(tiroTwo)
        tresMonedas.append(tiroThree)
        combinaciones.append(tresMonedas)

tiradas(numTiradas)
#print(combinaciones)

def checkProbability(arrayTiradas,numeroTiradas):
    
    ceroNums = [0,0,0]
    tresCeros = 0
    
    for i in arrayTiradas:
        
        if (i == ceroNums):
            tresCeros+=1
    
    porcentaje = (tresCeros*100)/numeroTiradas
    
    return porcentaje
    
    
        
checkProbability(combinaciones,numTiradas)

print("La probabilidad de que salgan 3 cruces seguidas en una tirada es de "+str(checkProbability(combinaciones,numTiradas))+"%")
```
